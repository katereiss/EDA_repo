```python
import pandas as pd
import numpy as np
```

```python
week_nums = [160903, 160910, 160917]
dfs = []
for week_num in week_nums:
  dfs.append(pd.read_csv(f"http://web.mta.info/developers/data/nyct/turnstile/turnstile_{week_num}.txt"))
turnstiles_df = pd.concat(dfs)
turnstiles_df.head()
```

```python
turnstiles_df.columns
```

```python
turnstiles_df.columns = turnstiles_df.columns.str.strip()
```

```python
turnstiles_df.columns = turnstiles_df.columns.str.lower()
```

```python
turnstiles_df.columns = turnstiles_df.columns.str.replace('/','_')
```

```python
turnstiles_df['datetime'] = pd.to_datetime(turnstiles_df['date'] + ' ' + turnstiles_df['time'])
```

```python
turnstiles_df.head()
```

```python
from datetime import datetime
```

```python
turnstiles_df['day_of_week'] = turnstiles_df['datetime'].dt.day_name()
turnstiles_df.head()
```

```python
turnstiles_df['daily_entries'] = turnstiles_df.entries.diff()
turnstiles_df['daily_exits'] = turnstiles_df.exits.diff()
turnstiles_df['daily_traffic'] = turnstiles_df.entries.diff() + turnstiles_df.exits.diff()
```

```python
turnstiles_daily = (turnstiles_df.groupby(['c_a','unit','scp','station','datetime'],as_index = False).daily_entries.first())
```
